```python
import numpy as np
import pandas as pd
df= pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.isnull().sum()  #check missing values
```

```python
# to replace null values either by mean or median.  but why?  usually median over mean as mean affected by outliers
#but it depends on situation.
# median isnt affected by outliers
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
# first find median reveiw rating in each category and then replace the null value with this median review rating for the specific category.
# saves integrity of the data and no misunderstanding
```

```python
df.isnull().sum()
```

```python
# better to keep all the column names in snake_case  as saves time and consitent and reference in python and sql
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df= df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# FEATURE ENGINEERING to create a new column
# age group to group cx into 4 age groups.

# pd.qcut does > splits ages into 4 equal sized grps based on data distirbution and assigns labels
```

```python
labels = ['Young Adult', 'Adult', 'Middle Aged', 'Senior']
df['age_group']= pd.qcut(df['age'], q=4, labels= labels)
```

```python
df[['age','age_group']].head(8)
```

```python
# anohter Feature;  purchase freq days

frequency_mapping = {
    'Fortnightly':14,
    'Weekly':7,
    'Monthly': 30,
    'Quarterly': 90,
    'Annually': 365
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
# created a new column which maps the text in column with no of days

df[['purchase_frequency_days', 'frequency_of_purchases']].head(10)
```
